# Load MNIST Dataset

```python
import numpy as np
from sklearn import datasets
mnist = datasets.fetch_openml('mnist_784', version=1, return_X_y = True)
X, y = mnist
n = int(np.ceil(np.sqrt(X.shape[1])))
```

# Select Collection of Two's from Dataset

```python
two = y[y=='2']
a=np.array(two.index,dtype='int32')
twos = X.iloc[a]
twos
```

# Fit a rank 25 NMF model to the collections of Two's

```python
import numpy as np
from sklearn.decomposition import NMF
model = NMF(n_components=25, init='random', random_state=42, max_iter=1000)
W = model.fit_transform(twos)
H = model.components_
```

# Fit a rank of 24 PCA to the collections of Two's

```python
from sklearn.decomposition import PCA
pca_24 = PCA(n_components=24)
pca = pca_24.fit_transform(twos)
```

# Display W and H matrics

```python
import matplotlib.pyplot as plt
print("\n\n=====================================================")
print("                       W Matrix")
print("=====================================================\n\n")
w=20
h=20
fig=plt.figure(figsize=(7, 5))
columns = 6
rows = 4
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(H[i].reshape(28, 28), cmap='gray')
    plt.xticks([]),  plt.yticks([])
```

```python
print("\n\n=====================================================")
print("                       H Matrix")
print("=====================================================\n\n")
w=20
h=20
fig=plt.figure(figsize=(7, 7))
columns = 7
rows = 7
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(W[i].reshape(5,5),cmap='gray')
    plt.xticks([]),  plt.yticks([])
```
